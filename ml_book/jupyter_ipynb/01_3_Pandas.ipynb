### ▶ DataFrame, Index, Series

(DataFrame은 여러 개의 Series로 이루어졌다고 할 수 있음)

#### dictionary to DataFrame & DataFrame to dictionary

```python
# Key는 컬럼명으로 매핑, Value는 리스트 형(또는 ndarray)
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

```python
# DataFrame to dictionary - DataFrame.to_dict(DataFrame)
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

### ▶ DataFrame의 column 데이터 세트 생성과 수정

```python
# 새 column 추가 및 값 할당
titanic_df['Age_0']=0
titanic_df.head(3)
```

```python
# 기존 column을 가공하여 새 column 추가
titanic_df['Age_by_10'] = titanic_df['Age']*10
titanic_df['Family_No'] = titanic_df['SibSp'] + titanic_df['Parch']+1
titanic_df.head(3)
```

```python
# 기존 column의 값 update
titanic_df['Age_by_10'] = titanic_df['Age_by_10'] + 100
titanic_df.head(3)
```

### ▶ DataFrame 데이터 삭제
#### axis 0 - row 방향 drop / axis 1 - col 방향 drop
#### 

```python
# 데이터 삭제 - drop()
titanic_drop_df = titanic_df.drop('Age_0', axis=1 )
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
```

```python
# inplace 속성 - 원본 Data Frame을 변경 - True / 변경하지 않음 - False
drop_result = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
print(' inplace=True 로 drop 후 반환된 값:',drop_result)
titanic_df.head(3)
```

```python
# 맨 앞 3개의 row 삭제해보기
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 15)
print('#### before axis 0 drop ####')
print(titanic_df.head(3))

titanic_df.drop([0,1,2], axis=0, inplace=True)

print('#### after axis 0 drop ####')
print(titanic_df.head(3))
```
