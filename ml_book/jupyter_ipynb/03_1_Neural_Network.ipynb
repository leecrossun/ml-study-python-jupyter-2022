```python
import numpy as np
import matplotlib.pylab as plt
def step_function(x):
    return np.array(x > 0, dtype=np.int64)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

###  Sigmoid vs Step

```python
# sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
print(sigmoid(x))
```

###  ReLU(Rectified Linear Unit) function
- ReLU 0     , 0  0

0  \;if\; x <= 0\end{cases}

```python
def relu(x):
    return np.maximum(0, x)
```
